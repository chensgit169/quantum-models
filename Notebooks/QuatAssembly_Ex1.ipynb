Consider the Hamiltonian 

$$
H = \sum^N_{n=1}\epsilon_n∣n⟩⟨n| 
+
\sum^{N−1}_{n=1} V (|n⟩⟨n+1| + |n+1⟩⟨n|)
$$ 
. For $\epsilon_n$ = ϵ and N = 20 determine the eigenvalues and eigenvectors. Calculate the
Probabilities $P_n(t) = |⟨n|\psi(t)⟩|^2$ and visualize the evolution.

```python
import sys
sys.path.append("../")
import numpy as np
```

```python
from hamiltonian.models.tight_binding import TightBinding
from hamiltonian.visualization.time_evolution import draw_heatmap
```

```python
# define model
n = 50
e = 2
u = 4

model = TightBinding(n, e, u)

psi_0 = np.zeros(n)
n_i = 0
psi_0[n_i] = 1  # |psi_0> = |1>
ts = np.linspace(0, 10, 5000) + 100

ps = [np.abs(model.psi_t(psi_0, t))**2 for t in ts]
ps = np.array(ps)

ax = draw_heatmap(ps, ts=ts)
```
