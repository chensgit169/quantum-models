```python
x = symbols('x')
Q = Function('Q')(x)  # General function Q(x)
P_list_general = compute_p_list(Q, x, 2, branch=1)
```

```python
from sympy import diff

Q1, Q2, Q3 = symbols("Q' Q'' Q'''")  # custom symbols with prime notation

expr = P_list_general[3]
expr_sub = expr.subs({
    diff(Q, x): Q1,
    diff(Q, x, 2): Q2,
    diff(Q, x, 3): Q3
})
expr_sub
```
